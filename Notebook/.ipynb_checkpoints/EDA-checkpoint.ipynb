```python
rows_per_file = 3000  
selected_dfs = []

for file in csv_files:
    df = pd.read_csv(file)
    
    if len(df) >= rows_per_file:
        selected_dfs.append(df.sample(n=rows_per_file, random_state=42))
    else:
        selected_dfs.append(df)  # take all rows if fewer


# Combining everything
combined_df = pd.concat(selected_dfs, ignore_index=True)

# Saving the file
combined_df.to_csv("Amazon_data.csv", index=False)

print("✅ Combined CSV created successfully!")
```

# Pipeline for Data Cleaning

```python
def preprocess_data(df) -> pd.DataFrame:
    """
    Preprocess product dataset for recommendation system.
    Handles:
    - Price cleaning
    - Ratings and no_of_ratings conversion
    - Null value handling
    - Standardization of column names
    """
    
    # ---- 1. Normalize column names ----
    df.columns = [col.strip().lower().replace(" ", "_") for col in df.columns]

    # ----2. Clean Price data ----- 
    for col in ['discount_price','actual_price']:
        df[col] = ( 
            df[col]
            .astype(str)
            .str.replace("₹","", regex=False)
            .str.replace(",","", regex=False)
            .replace("nan", np.nan)
            .astype(float)
            )
        
    # ----3. Clean Ratings Data ----
    df['ratings'] = pd.to_numeric(df['ratings'], errors='coerce')
    df['ratings'] = df['ratings'].fillna(df['ratings'].mean())

    # ----4. Cleaning Number_of_Rating ----
   
    df['no_of_ratings'] = (
        df['no_of_ratings']
        .astype(str)                         
        .str.replace(",", "", regex=True)     
    )

    df["no_of_ratings"] = pd.to_numeric(df["no_of_ratings"], errors="coerce")
    df["no_of_ratings"] = df["no_of_ratings"].fillna(0).astype(int)

    # ---- 5. Filling the remaining missing numeric values ----
    for col in ["discount_price", "actual_price"]:
        if col in df.columns:
            df[col].fillna(df[col].median(), inplace=True)
    
    
    return df
```

```python
df = preprocess_data(df)
```
